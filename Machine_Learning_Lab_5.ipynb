# **Linear Regression**

### 1.1 Linear Regression Example

```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Step 2: Define the data
# X = CGPA
# Y = Salary in LPA
X = np.array([6, 7, 8, 9, 10]).reshape(-1, 1)
Y = np.array([3, 4, 5, 7, 8])
# Step 3: Create Linear Regression model
model = LinearRegression()
# Step 4: Training the model
model.fit(X, Y)
# Step 5: Get model parameters
slope = model.coef_[0]
intercept = model.intercept_
print("Slope (m):", slope)
print("Intercept (c):", intercept)
# Regression equation
print(f"Regression Equation: Y = {slope:.2f}X + {intercept:.2f}")
# Step 6: Predict values
y_pred = model.predict(X)
print("Predicted values:", y_pred)

# Step 7: Predict salary for new GPA
new_gpa = np.array([[8.5]])
predicted_salary = model.predict(new_gpa)
print("Predicted salary for GPA 8.5:", predicted_salary[0])

# Step 8: Plot the data and regression line
plt.scatter(X, Y)          # actual data
plt.plot(X, y_pred)        # regression line
plt.xlabel("GPA")
plt.ylabel("Salary (LPA)")
plt.title("Linear Regression: GPA vs Salary")
plt.show()
```

### 1.2 Linear Regression on House Price Prediction Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Load the dataset
df_house = pd.read_csv('House Price Prediction Dataset.csv')

# Display the first few rows and info to understand the data
print("House Price Dataset Head:")
display(df_house.head())
print("\nHouse Price Dataset Info:")
df_house.info()
```

```python
# Step 2: Define features (X) and target (Y)
# For simplicity, let's use 'Area' as the independent variable (X) and 'price' as the dependent variable (Y)
X_house = df_house['Area'].values.reshape(-1, 1) # Reshape for sklearn
Y_house = df_house['Price'].values

# Step 3: Create Linear Regression model with no intercept
model_house = LinearRegression(fit_intercept=False)

# Step 4: Training the model
model_house.fit(X_house, Y_house)

# Step 5: Get model parameters
slope_house = model_house.coef_[0]
intercept_house = model_house.intercept_
print(f"\nSlope (m) for House Price: {slope_house:.2f}")
print(f"Intercept (c) for House Price: {intercept_house:.2f}")

# Regression equation
print(f"Regression Equation: Y = {slope_house:.2f}X + {intercept_house:.2f}")

# Step 6: Predict values
y_pred_house = model_house.predict(X_house)

# Step 7: Plot the data and regression line
plt.figure(figsize=(10, 6))
plt.scatter(X_house, Y_house, alpha=0.3, label='Actual Data') # actual data
plt.plot(X_house, y_pred_house, color='blue', label='Regression Line') # regression line
plt.xlabel("Square Footage (Area)")
plt.ylabel("Price")
plt.title("Linear Regression: Area vs. Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def gradient_descent(x, y):
    m_curr = 0
    b_curr = 0
    learning_rate = 0.08
    iterations = 1000
    n = len(x)

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr

        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])

        dm = (-2/n) * sum(x * (y - y_predicted))
        db = (-2/n) * sum(y - y_predicted)

        m_curr = m_curr - learning_rate * dm
        b_curr = b_curr - learning_rate * db

        print("m {}, b {}, cost {}, iteration {}".format(m_curr, b_curr, cost, i))

    return m_curr, b_curr

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
m, b = gradient_descent(x, y)
print("Final values:")
print("Slope (m):", m)
print("Intercept (b):", b)
```
