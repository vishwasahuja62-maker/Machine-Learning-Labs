```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load the in-built tips dataset from seaborn
dataset = sns.load_dataset('tips')
display(dataset.head())
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='total_bill', y='tip', sizes=(20, 400), alpha=0.7)
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='total_bill', y='tip', hue='sex', sizes=(20, 400), alpha=0.7)
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='total_bill', y='tip',hue='sex', size='size', sizes=(20, 400), alpha=0.7)
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='total_bill', y='tip',hue='sex', size='size', style='smoker', sizes=(20, 400), alpha=0.7)
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load the in-built tips dataset from seaborn
dataset = sns.load_dataset('titanic')
display(dataset.head())
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=dataset, x='pclass', y='survived', hue='pclass', palette='viridis', legend=False)
plt.title('Survival Rate by Passenger Class (Pclass)')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load the in-built tips dataset from seaborn
dataset = sns.load_dataset('flights')
display(dataset.head())
```

```python
flights_pivot = dataset.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt='.0f', cmap='viridis', linewidths=.5)
plt.title('Passenger Count per Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
sns.pairplot(dataset, hue='month', palette='viridis')
plt.suptitle('Pair Plot of Flights Dataset', y=1.02) # Add a title for the whole plot
plt.show()
```

```python
sns.clustermap(flights_pivot, cmap='viridis', figsize=(10, 10))
plt.suptitle('Cluster Map of Flights Dataset (Passengers by Month and Year)', y=1.02)
plt.show()
```
