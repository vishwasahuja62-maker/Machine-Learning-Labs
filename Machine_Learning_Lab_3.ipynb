```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Data Frams elements can be heterogeneous
df = pd.read_csv('Titanic-Dataset.csv')
display(df.head())
```

```python
# Countplot
plt.figure()
sns.countplot(x='Survived', data=df)
plt.title('Count Plot of Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Pie Chart
plt.figure()
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of Survival')
plt.ylabel('')
plt.show()
```

```python
# Histogram
plt.figure()
plt.hist(df['Age'].dropna(), bins=20)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distplot
plt.figure()
sns.histplot(df['Fare'], kde=True)
plt.title('Distribution Plot of Fare')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

```python
# BoxPlot
plt.figure()
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Boxplot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
