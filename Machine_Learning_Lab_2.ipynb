```python
import pandas as pd
import numpy as np
# Data Frams elements can be heterogeneous
dataset = pd.read_csv('Salary_Dataset.csv')
display(dataset.head())
```

```python
# Applying One-Hot Encoding
country_dummy = pd.get_dummies(dataset['country'])
```

```python
dataset = pd.concat([dataset, country_dummy], axis = 1)
```

```python
dataset.drop('country', axis = 1, inplace = True)
```

```python
dataset = dataset[['Australia', 'Canada', 'Dubai', 'USA',
                   'YearsExperience', 'Salary', 'Purchased']]
dataset
```
