```python
import pandas as pd
import numpy as np
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
# Data Frams elements can be heterogeneous
df = pd.read_csv('car_dataset.data', names=headers)
# Using head() method to display first five rows of the dataset
display(df.head())
print(df.describe(include="all"))
print(df.info())
df.replace('?',np.nan, inplace=True)
print(df)
df['normalized-losses'].isnull().sum()
```

```python
# Creating a dataframe showing missing values
missing_data = df.isnull()
# Display first 5 rows of the missing data indicator
missing_data.head(5)
```

```python
# Counting missing values in each frame using 'for' loop
for column in headers:
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
# Alternative loop format with consistent spacing
for column in headers:
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

Based on the analysis, each column has 205 rows of data.
Seven columns contain missing data:

1. "normalized-losses": 41 missing values
2. "num-of-doors": 2 missing values
3. "bore": 4 missing values
4. "stroke": 4 missing values
5. "horsepower": 2 missing values
6. "peak-rpm": 2 missing values
7. "price": 4 missing values

Strategy to Deal with Missing Data:
- For numerical columns: Replace with mean/average
- For categorical columns: Replace with most frequent value (mode)
- For critical columns: Drop rows with missing values

```python
# Calculate the mean of normalized-losses column
# First convert to float type as data might be stored as strings
avg_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
print("Average of normalized-losses: ", avg_norm_loss)
```

```python
# Replace NaN values in normalized-losses with the calculated mean
# This fills missing values with the average value
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Replace missing 'num-of-doors' values with the most frequent value
# In this dataset, "four" is the most common number of doors
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# Dropping the whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)
# Resetting the index after dropping rows
df.reset_index(drop=True, inplace=True)
# Verifying the final state of missing data
missing_data = df.isnull()
missing_data.head(5)
```

```python
# Lets first change the data type of horsepower because I forgot to include it in to_float list of upper cell.
df["horsepower"] = df["horsepower"].astype(float)
```
